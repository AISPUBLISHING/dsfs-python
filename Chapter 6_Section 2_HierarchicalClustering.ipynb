## Import the libraries

```python
import numpy as np
import pandas as pd
from scipy.cluster.hierarchy import linkage, dendrogram
from sklearn.preprocessing import normalize
import matplotlib.pyplot as plt
import os
```

## Fix the Path

```python
os.chdir('D:')
os.getcwd()
```

## Import the dataset

```python
movements=pd.read_csv('stock_movements.csv',index_col=0)
```

## Normalize the dataset

```python
normalized_movements = normalize(movements)
```

## Calculate the linkage: margings Complete

```python
margings = linkage(normalized_movements,method='complete')
plt.figure(figsize=(22,10))
dendrogram(margings,
           labels=movements.index.values,
           leaf_rotation=80,
           leaf_font_size=12)
plt.show()
```

## Calculate the linkage: margings Single

```python
margings = linkage(normalized_movements,method='single')
plt.figure(figsize=(22,10))
dendrogram(margings,
           labels=movements.index.values,
           leaf_rotation=80,
           leaf_font_size=12)
plt.show()
```

## Calculate the linkage: margings Average

```python
mergings = linkage(normalized_movements,method='average')
plt.figure(figsize=(22,10))
dendrogram(mergings,
           labels=movements.index.values,
           leaf_rotation=80,
           leaf_font_size=12)
plt.show()
```
